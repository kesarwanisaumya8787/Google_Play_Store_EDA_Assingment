```python
def size_proces(item):
    if str(item)[-1]=="M":
        res=float(str(item).replace("M", ""))
        return res*1024
    elif str(item)[-1]=="k":
        return float(str(item).replace("k", ""))
    else:
        return str(np.nan)
    
```
